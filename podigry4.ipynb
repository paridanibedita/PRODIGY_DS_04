```python
df = data.drop_duplicates()
df
```

```python
##check null values
df.isnull().sum()
```

```python
##remove null values
cleandata = df.dropna()
cleandata.isnull().sum()
```

```python
###check unique values of sentimate column
sentimate_counts = cleandata['sentimate'].value_counts()
sentimate_counts
```

```python
##plot a count plot
sentimate_counts.plot(kind = 'bar')
plt.title('sentimate counts')
plt.xlabel('sentimates')
plt.ylabel('counts')
```

```python
brand_sentiment_counts = df[df['company'].str.contains('Microsoft', case=False)]['sentimate'].value_counts()
```

```python
brand_sentiment_counts.plot(kind = 'bar')
plt.title('sentimates of people for microsoft')
```

```python
# List of companies to loop through
companies = ['Microsoft', 'Nvidia', 'Amazon','TomClancysGhostRecon','Cyberpunk2077','johnson&johnson','MaddenNFL','Facebook','CallOfDuty'
            ,'FIFA','HomeDepot','Verizon','Hearthstone','Fortnite','LeagueOfLegends','ApexLegends','AssassinsCreed','Google','CS-GO','WorldOfCraft']

for company in companies:
    # Filter data for each company and count sentiments
    brand_sentiment_counts = df[df['company'].str.contains(company, case=False)]['sentimate'].value_counts()

    # Plot the sentiment counts as a bar chart
    brand_sentiment_counts.plot(kind='bar')
    
    # Set the title of the plot
    plt.title(f'Sentiments of people for {company}')
    
    # Show the plot
    plt.show()
```
